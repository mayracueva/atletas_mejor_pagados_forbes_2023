```python
import pandas as pd
df = pd.read_excel('/content/drive/MyDrive/Data/The_Worlds_Highest_Paid_Athletes.xlsx')
```

```python
df.head()
```

```python
df.info()
```

```python
df.rename(columns = {'Rank' : 'rank',
                     'Name' : 'name',
                     'Sport' : 'sport',
                     'Total Earnings' : 't_earnings',
                     'On-the-Field Earnings' : 'on_the_field',
                     'Off-the-Field Earnings' : 'off_the_field'}, inplace = True)
df.head()
```

```python
df[['on_the_%']] = None
df['on_the_%'] = df['on_the_field'] / df['t_earnings'] * 100
df[['off_the_%']] = None
df['off_the_%'] = df['off_the_field'] / df['t_earnings'] * 100
df.head()
```

```python
df.describe().T
```

```python
import plotly.express as px
fig = px.scatter(df, x="on_the_%", y="t_earnings", size="t_earnings", color="sport", hover_name="name",
                 labels={'on_the_%' : 'Ganancias en el campo (%)', 't_earnings' : 'Ganancias Totales ($)', 'sport' : '<i>Deporte<i>'},
                 size_max=40)
fig.update_xaxes(title_font=dict(size=18), title_standoff = 10,
                 tickfont=dict(size=14), tickangle=0, ticklabelposition='outside')
fig.update_yaxes(title_font=dict(size=18), title_standoff = 0,
                 tickfont=dict(size=14), tickangle=0, ticklabelposition='outside')
fig.update_legends(title_font=dict(size=15),
                   y=1.02, yanchor="top",
                   x=1.12, xanchor="right",
                   orientation="v")
fig.update_layout(font_family="Arial",
                  font_color="black",
                  title={'text':"<b>Los atletas mejor pagados del mundo</b><br><sup>Top 50 Forbes 2023<sup>", 'font':dict(size=20), 'automargin':True, 'yref':'container',
                         'y':0.94,
                         'x':0.5,
                         'xanchor': 'center',
                         'yanchor': 'top'})
fig.show()
```

```python
fig.write_html("plotly.html")
```
